```python
from bs4 import BeautifulSoup
from urllib.request import urlopen,Request
import concurrent.futures
```

```python
results  = []
url = "https://www.baaa-acro.com/crash-archives?page="

def crawl(page_number):
    urls = []
    print("Scraping page " + str(page_number) + " of 1722")
    page = urlopen(Request(url + str(page_number), headers={'User-Agent': 'Mozilla/5.0'}))
    html = page.read().decode("utf-8")
    soup = BeautifulSoup(html, "html.parser")
    for a in soup.find_all("a", class_="red-btn"):
        urls.append(a.get("href"))
    return urls
    
    
with concurrent.futures.ThreadPoolExecutor(max_workers=10) as executor:
    # Use a single for loop to submit tasks to the executor
    for i in range(0, 1723):
        future = executor.submit(crawl, i)
        results.append(future)

result_values = [result for future in results for result in future.result()]  
```

```python
import pandas as pd
df = pd.DataFrame(result_values, columns=["url"])
df.head(50)
```
