```python
import pandas as pd
```

```python
df = pd.read_csv('./src/AccidentesAviones.csv', na_values='?')
```

```python
df.tail()
```

### Eliminamos algunas columnas y renombramos otras

```python
df.drop(['Unnamed: 0', 'HORA declarada', 'flight_no', 'registration', 'cn_ln'], axis=1, inplace=True)
```

```python
df.rename(columns={'fecha': 'Fecha', 'Ruta': 'Ubicación', 'OperadOR': 'Operador', 'route': 'Ruta', 'ac_type': 'Tipo de aeronave', 'all_aboard': 'Total a bordo',\
    'PASAJEROS A BORDO': 'Pasajeros a bordo', 'crew_aboard': 'Tripulación a bordo', 'cantidad de fallecidos': 'Total fallecidos', 'passenger_fatalities': 'Fallecidos pasajeros',\
    'crew_fatalities': 'Fallecidos tripulación', 'ground': 'Fallecidos en tierra', 'summary': 'Resumen'}, inplace=True)
```

### Guardamos el CSV limpio

```python
df.to_csv('./src/AccidentesAvionesLimpio.csv', index=False)
```
